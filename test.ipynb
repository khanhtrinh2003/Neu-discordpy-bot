```python
import re
```

```python
message = "Câu 0\n cau 1 \n cau2 \n cau 3\nCâu 0\n cau 1 \n cau2 \n cau 3\nCâu 0\n cau 1 \n cau2 \n cau 3\nCâu 0\n cau 1 \n cau2 \n cau 3\nCâu 0\n cau 1 \n cau2 \n cau 3"
```

```python
def rut(message):
    mes = message.split("\n")
    kqs = []
    for i in range(1,len(mes)):
        hop = "\n".join(mes[(i-1)*5:i*5])
        kqs.append(hop)
    return kqs
```

```python
a = rut(message)
while("" in a):
    a.remove("")
```
